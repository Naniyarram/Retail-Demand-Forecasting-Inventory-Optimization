```python
import pandas as pd
import numpy as np

# Create new time series data (e.g., February 2023)
data = {
    'date': pd.date_range(start='2025-02-01', periods=10, freq='D'),
    'value': [12, 18, 21, 17, 25, 29, 33, 28, 24, 20]
}
df = pd.DataFrame(data)

df
```

```python
# Set the date column as the index
df.set_index('date', inplace=True)
df
```

```python

# Calculate the moving average with a window size of 4

df['rolling_mean_4'] = df['value'].rolling(window=4).mean()
df
```

```python
# use the above time series data sample
df['rolling_std_4'] = df['value'].rolling(window=4).std()

print(df)
```

```python
df['rolling_expanding_mean'] = df['value'].expanding().mean()
df
```

```python
df['running_total'] = df['value'].expanding().sum()
df
```
